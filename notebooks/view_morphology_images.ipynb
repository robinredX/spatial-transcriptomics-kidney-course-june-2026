```python
from tifffile import imread
```

```python
import matplotlib.pyplot as plt
```

```python
im = imread("shared/MBI_BIA_Project_data/Xenium_data/morphology.ome.tif", level=2) # 16 channels 
im_focus = imread("shared/MBI_BIA_Project_data/Xenium_data/morphology_focus.ome.tif", level=2) # 2D
```

```python
## Invert colors - to make nuclei black and background white
import numpy as np
im_invert = np.iinfo(im_focus.dtype).max - im_focus
```

```python
plt.imshow(im_invert, cmap="gray")
plt.title("2D collapsed - Inverted version (Nuclei in Black)")
```

```python
plt.imshow(im_focus, cmap="gray")
plt.title("2D collapsed")
```

```python
for ch in range(im.shape[0]):
    plt.imshow(im[ch], cmap="gray")
    plt.title(ch)
    plt.show()
```
